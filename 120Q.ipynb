```python
def vowelConsonant(string):
  vowels='aeiouAEIOU'
  vow=''
  con=''
  string=string.replace(" ","")  #Removes spaces
  for char in string:
    if char in vowels:
      vow+=char
    else:
      con+=char
  return (f'{vow}\n{con}')
print(vowelConsonant("O NITHIN SAI BALAJI"))
```

17. Write a program to find the smallest of three numbers.

```python
def smallestOfThreeNumbers(a16, b16, c16):

    if a16 == b16 == c16:
        return "All three numbers are equal"

    smallest = min(a16, b16, c16)

    # Two equal smallest
    if a16 == b16 and a16 < c16:
        return f"{a16} and {b16} are equal and smallest"
    elif b16 == c16 and b16 < a16:
        return f"{b16} and {c16} are equal and smallest"
    elif c16 == a16 and c16 < b16:
        return f"{c16} and {a16} are equal and smallest"

    # Single smallest
    return f"{smallest} is the smallest number."
print(smallestOfThreeNumbers(45, 7, 18))
print(smallestOfThreeNumbers(18, 26, 5))
print(smallestOfThreeNumbers(18, 7, 45))

print(smallestOfThreeNumbers(45, 45, 18))
print(smallestOfThreeNumbers(9, 23, 23))
print(smallestOfThreeNumbers(31, 23, 31))

print(smallestOfThreeNumbers(23, 23, 23))
print(smallestOfThreeNumbers(45, 99, 45))
```

26. Write a program to check if a string is a palindrome.

```python
def palindrome(string26):
  left=0
  right=len(string26)-1
  while left<right:
    if string26[left]!=string26[right]:
      return "Not a palindrome"
    left+=1
    right-=1
  return "Palindrome"
print(palindrome('nitin'))
print(palindrome('nithin'))
```

```python
def palindrome(string26):
  if string26==string26[::-1]:  #Inbuilt function which reverse the string #only strings
    return ('palindrome')
  else:
    return ('Not a Palindrome')
print(palindrome('nitin'))
print(palindrome('nithin'))
```

28. Write a program to calculate the LCM of two numbers.

```python
def gcdFun(a,b):  # Function to find GCD
  while b!=0:
    a,b=b, a%b
  return a

def lcmFun(a,b):   #Function to find LCM
  if a<=0 or b<=0:
    return "Enter the number greater than 0"
  gcdVal=gcdFun(a,b)
  lcmVal = (a*b) // gcdVal
  return lcmVal


a=15
b=45
print(f'GCD: {gcdFun(a,b)}')
print(f'LCM: {lcmFun(a,b)}')
```

30. Write a program to print all prime numbers between 1 and 100.

```python
def PrintPrimeNumbers(num):
  for num in range(1,num+1):
    isPrime=True
    return "Enter the positive number"
  for i in range (2, int(num**0.5)+1):
    if num%i==0:
      isPrime=False
  if isPrime:
    print(num)

print(PrintPrimeNumbers(100))
```
